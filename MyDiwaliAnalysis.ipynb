# Exploratory Data Analysis

```python
df.columns
```

Gender Count

```python
ax=sns.countplot(x='Gender',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

Gender vs AmountPlot

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
temp=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for i in temp.containers:
    temp.bar_label(i)
```

Age Count

```python
sorted_age_groups = sorted(df["Age Group"].unique(), key=lambda x: int(x.split('-')[0]) if '-' in x else int(x[:-1]))

ax=sns.countplot(data=df,x="Age Group",hue="Gender", order=sorted_age_groups)
for i in ax.containers:
    ax.bar_label(i)
```

Age vs Amount Plot

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
temp=sns.barplot(x='Age Group',y='Amount',data=sales_age)
for i in temp.containers:
    temp.bar_label(i)
    
```

State count

```python
df.columns
```

```python
ax=sns.countplot(x="State",data=df)
sns.set(rc={'figure.figsize':(16,5)})
for i in ax.containers:
    ax.bar_label(i)
```

State vs Orders plot

```python
sales_states=df.groupby(["State"], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
temp=sns.barplot(data=sales_states, x='State',y='Orders')
for i in temp.containers:
    temp.bar_label(i)
```

State vs Amount

```python
sales_states=df.groupby(["State"], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
temp=sns.barplot(data=sales_states, x='State',y='Amount')
for i in temp.containers:
    temp.bar_label(i)
```

Martial Status

```python
ax=sns.countplot(x="Marital_Status",data=df)
sns.set(rc={'figure.figsize':(4,5)})

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales=df.groupby(["Marital_Status","Gender"], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
temp=sns.barplot(data=sales, x='Marital_Status',y='Amount',hue="Gender")
for i in temp.containers:
    temp.bar_label(i)
    
```

Occupation Count

```python
ax=sns.countplot(x="Occupation",data=df)
sns.set(rc={'figure.figsize':(25,5)})

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales=df.groupby(["Occupation"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

temp=sns.barplot(data=sales,x="Occupation",y='Amount')

for i in temp.containers:
    temp.bar_label(i)
```

Product Category

```python
ax=sns.countplot(x="Product_Category",data=df)
sns.set(rc={'figure.figsize':(15,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(["Product_Category"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

temp=sns.barplot(data=sales,x="Product_Category",y='Amount')

for i in temp.containers:
    temp.bar_label(i)
```

```python
sales=df.groupby(["Product_Category"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

temp=sns.barplot(data=sales,x="Product_Category",y='Orders')

for i in temp.containers:
    temp.bar_label(i)
```

```python
sales=df.groupby(["Product_ID"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

temp=sns.barplot(data=sales,x="Product_ID",y='Orders')

for i in temp.containers:
    temp.bar_label(i)
```
